```python
import matplotlib.pyplot as plt
import random as rnd
import numpy as np
```

```python
d, s = 100, 100 # Anzahl betrunkener, Anzahl Schritte
distlist = [] # Liste der quadratischen Entfernungen jedes Betrunkenen nach s Schritten
```

```python
for i in range(d):
    slist = [(0,0)] # Schrittliste
    x = 0
    for j in range(s):
        x += (-1)**rnd.randint(0, 1)
        slist.append((j+1, x))
    distlist.append(slist[s][1]**2)
    x,y = zip(*slist)
    plt.plot(x,y)
plt.show()
```

```python
rms = (sum(distlist)/len(distlist))**0.5
print(rms)
```

## Im 2d Fall

```python
for i in range(d):
    slist = [(0,0)] # Schrittliste
    x, y = 0, 0
    for j in range(s):
        phi = 2*np.pi*rnd.random()
        x += np.cos(phi)
        y += np.sin(phi)

        slist.append((x, y))
    distlist.append((slist[s][0]**2+slist[s][1]**2)**0.5)
    x,y = zip(*slist)
    plt.plot(x,y)
plt.show()
```
